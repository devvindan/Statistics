## Шаг 1. Экспорт данных в систему обработки

Для анализа и визуализации данных используется язык программирования Python, а также библиотеки Numpy, Pandas и Matplotlib.

Первым делом подключим необходимые библиотеки и установим настройки локализации:

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

default_dpi = mpl.rcParamsDefault['figure.dpi']
mpl.rcParams['figure.dpi'] = default_dpi * 0.9

import pandas as pd
import numpy as np
import sys
import math
import scipy.stats as sts
from collections import Counter
```

Исходные данные хранятся в CSV-таблице размером 626x8. 

```python
data = pd.read_csv("honeyproduction.csv", header=0, sep=",")
data
```

Для анализа мы выберем 3 столбца:

- state - имя штата-производителя
- totalprod - общий вес произведённого мёда (в фунтах)
- priceperlb - цена за фунт (в $)

```python
n = data.shape[0] # n - количество строк в таблице (вариант)
state = data['state']
totalprod = data['totalprod']
priceperlb = data['priceperlb']
pd.DataFrame({"Название штата": state, "Всего произведено": totalprod, "Цена за фунт ($)": priceperlb})
```

Для работы с данными о штатах закодируем их названия числами:

```python
state = state.astype('category').cat.codes
pd.DataFrame({"Название штата": state, "Всего произведено": totalprod, "Цена за фунт ($)": priceperlb})
```

## Шаг 2. Анализ данных

Для начала определим классы данных:

Код штата принимает значение из определённого набора и является дискретными данными, в то время как цена за фунт (в $) и общее количество произведённого мёда может принять любое значение из промежутка и является непрерывными данными. 

Начнём с анализа кодов штата. Первым делом составим вариационный ряд, посчитаем размах и составим таблицу частот:

```python
frequencies_dict = Counter(state)
values = np.array(list(frequencies_dict.keys()))
values.sort()
frequencies = np.array([frequencies_dict[key] for key in values])
rel_frequencies = np.divide(frequencies.astype(float), n)
vrange = np.ptp(values)
print("Размах (максимальное - минимальное значение): {}".format(vrange))
```

Таблица частот:

* n - частота 
* w - относительная частота
* x - величина

```python
pd.DataFrame({"n" : frequencies, "x": values, "w" : rel_frequencies})
```

Построим график зависимости частоты w(i) от величины x:

```python
plt.plot(values, rel_frequencies)
plt.title("Полигон частот:")
plt.xlabel("x(i)")
plt.ylabel("w(i)")
plt.show()
```

Следующим шагом посчитаем и построим функцию распределения и построим её график: 

$$F(x) = P(X < x)$$

```python
cum_frequencies = [sum(rel_frequencies[0:index]) for index, value in enumerate(rel_frequencies)]
pd.DataFrame({"x":values, "w":rel_frequencies, "F(x)":cum_frequencies })
```

```python
plt.plot(values, cum_frequencies, color='red')
plt.title("График функции распределения:")
plt.xlabel("x(i)")
plt.ylabel("F(x)")
plt.show()
```

Посчитаем основные характеристики:

Математическое ожидание: $$M[X] = \frac{\sum_{i=1}^{n}x_i}{n}$$
или
$$M[X] = \frac{\sum_{i=1}^{k}x_i^{'} n_i}{n}$$

```python
expectancy = sum(state) / n
print("Математическое ожидание = {}".format(expectancy))
```

Дисперсия:
$$D[X] = \sigma^{2} = \frac{\sum_{i=1}^{n}(x_i - M[X])^{2}}{n}$$
или
$$D[X] = \sigma^{2} = \frac{\sum_{i=1}^{k}(x_i^{'} - M[X])^{2}n_i}{n}$$

```python
variance = sum([(x - expectancy) ** 2 for x in state]) / n
print("Дисперсия = {}".format(variance))
```

Среднее квадратичное отклонение:
$$\sigma = \sqrt{D[X]}$$

```python
standard_deviation = np.sqrt(variance)
print("Среднее квадратичное отклонение = {}".format(standard_deviation))
```

```python
median = np.median(state)
print("Медиана = {}".format(median))
```

Исходя из визуального представления функций распределения и частот выдвинем гипотезу, что данное распределение является равномерным. 

Теоретические частоты:

$$m^{'} = \frac{1}{n}$$

где n - число уникальных вариант

```python
theor_frequencies = [1.0 / float(len(values)) for x in values]
pd.DataFrame({"m": rel_frequencies, "$m^{'}$": theor_frequencies})
```

Для проверки на равномерное распределение вычислим критерий Пирсона:

$$\chi^{2} = \sum\frac{(m_i - m_i^{'})^{2}}{m_i^{'}}$$

```python
xi_squared = sum([((rel_frequencies[i] - theor_frequencies[i]) ** 2) 
                  / theor_frequencies[i] for i in range(len(rel_frequencies)) ])
print("Критерий Пирсона: {}".format(xi_squared))
```

Число степеней свободы k = s - 3, где s - количество разных вариант.

```python
k = len(values) - 3
print("Число степеней свободы k: {}".format(k))
```

Сравнив полученные результаты для $\chi^{2}$ с табличными, сделаем вывод, что полученное значение хи-квадрат значительно меньше табличных. Следовательно, нет основания отвергать гипотезу.

Далее проведём расчёты для общего веса произведённого мёда в фунтах:

```python
frequencies_dict = Counter(totalprod)
values = np.array(list(frequencies_dict.keys()))
values.sort()
frequencies = np.array([frequencies_dict[key] for key in values])
rel_frequencies = np.divide(frequencies.astype(float), n)
vrange = np.ptp(values)
print("Размах (максимальное - минимальное значение): {}".format(vrange))
```

Определим шаг по формуле Стерджесса
$$h = \frac{x_{max} - x_{min}}{1 + 3,22\lg n}$$

```python
min_value = np.amin(totalprod)
max_value = np.amax(totalprod)
vrange = max_value - min_value
h = (max_value - min_value) / (1 + 3.22 * np.log10(n)) # шаг
k = int(np.floor((max_value - min_value) / h)) # количество интервалов
frequencies_dict = Counter(totalprod)
values = np.array(list(frequencies_dict.keys()))
values.sort()
totalprod = np.array(totalprod)
print("Минимальное значение: {} Максимальное значение: {} Шаг: {} k: {}".format(min_value, max_value, h, k))
```

Вычислим основные характеристики:

```python
expectancy = np.mean(totalprod)
variance = np.var(totalprod)
standard_deviation = np.std(totalprod)
median = np.median(totalprod)
print("Мат.ожидание: {}\nДисперсия: {}\nСтанд.отклонение: {}\nМедиана: {}\n"
      .format(expectancy, variance, standard_deviation, median))
```

Построим таблицу и график распределения:

```python
ints, bins, patches = plt.hist(totalprod, normed=True)
mids = [(bins[i] + bins[i + 1]) / 2 for i in range(k)]
rel_frequencies_group = np.divide(ints, n)

plt.show()
```

```python
pd.DataFrame({"n":ints, "$x_a, x_b$":["[ {} ; {} ]"
                                   .format(bins[i], bins[i + 1]) for i in range(k)],
              "$w_i$":rel_frequencies_group, "$x_i$":mids})
```

Проведём анализ разбиения на группы, для чего вычислим:
    
Межгрупповую дисперсию:

$$\sigma_{мг}^{2} = \frac{\sum_{i=1}^{k} (\tilde{x_i} - \tilde{x_{общ}})^{2} * f_i}{\sum_{i=1}^{k}f_i}$$

```python
mean = np.mean(totalprod)
#межгрупповая дисперсия
intergroup_var = sum([((np.mean(totalprod[np.where((totalprod > bins[i]) & (totalprod <= bins[i + 1]))]) 
                    - mean) ** 2) * rel_frequencies_group[i] for i in range(k)]) / sum(rel_frequencies_group)
print("Межгрупповая дисперсия: {}".format(intergroup_var))
```

Внутригрупповые дисперсии:

$$\sigma_{j}^{2} = \frac{\sum_{i=1}^{n}(x_i - \tilde{x_j})^{2}}{n}$$

```python
#средняя из внутригрупповых дисперсий
ingroup_var = np.mean([np.var(totalprod[np.where((totalprod > bins[i]) & (totalprod <= bins[i + 1]))])
                       for i in range(k)])
print("Средняя из внутригрупповых дисперсий: {}".format(ingroup_var))
```

По правилу сложения дисперсий, дисперсия равна сумме средней из внутригрупповых дисперсий и межгрупповой дисперсии:
$$\sigma^{2} = \sigma_{мг}^{2} + \tilde{\sigma^{2}}$$

```python
print("Дисперсия: {}\nСумма: {}".format(variance, ingroup_var + intergroup_var))
print("Различие: {} %".format(np.abs(variance - (ingroup_var + intergroup_var)) / variance * 100))
```

Построим функцию распределения для $x_i$:

```python
cum_frequencies = [sum(rel_frequencies_group[0:index]) for index, value in enumerate(rel_frequencies_group)]
plt.plot(mids, cum_frequencies, color='red')
plt.title("График функции распределения:")
plt.xlabel("x(i)")
plt.ylabel("F(x)")
plt.show()
```

Проанализировав графики, выдвинем гипотезу, что данное распределение является экспоненциальным (показательным). Для проверки гипотезы используем критерий согласия Пирсона.

Вычислим теоретические частоты по формуле для экспоненциального распределения при x > 0:

$$f(x) = \lambda e^{-\lambda x}$$

где $$\lambda = \frac{1}{\tilde{x}}$$

```python
lmbd = 1.0 / expectancy
theor_frequencies = [lmbd * np.exp(-1 * lmbd * x) for x in mids]
```

Для проверки на равномерное распределение вычислим критерий Пирсона:

$$\chi^{2} = \sum\frac{(m_i - m_i^{'})^{2}}{m_i^{'}}$$

```python
xi_squared = sum([((rel_frequencies_group[i] - theor_frequencies[i]) ** 2) 
                  / theor_frequencies[i] for i in range(len(rel_frequencies_group)) ])
print("Критерий Пирсона: {}".format(xi_squared))
```

Полученное значение гораздо больше табличного, значит, гипотеза опровергается.

Далее проведём анализ цены за фунт (непрерывной величины):

```python
frequencies_dict = Counter(priceperlb)
values = np.array(frequencies_dict.keys())
values.sort()
frequencies = np.array([frequencies_dict[key] for key in values])
rel_frequencies = np.divide(frequencies.astype(float), n)
vrange = np.ptp(values)
print("Размах (максимальное - минимальное значение): {}".format(vrange))
```

Определим шаг по формуле Стерджесса
$$h = \frac{x_{max} - x_{min}}{1 + 3,22\lg n}$$

```python
min_value = np.amin(priceperlb)
max_value = np.amax(priceperlb)
vrange = max_value - min_value
h = (max_value - min_value) / (1 + 3.22 * np.log10(n)) # шаг
k = int(np.floor((max_value - min_value) / h)) # количество интервалов
frequencies_dict = Counter(priceperlb)
values = np.array(frequencies_dict.keys())
values.sort()
priceperlb = np.array(priceperlb)
print("Минимальное значение: {} Максимальное значение: {} Шаг: {}".format(min_value, max_value, h))
```

Построим таблицу и график распределения:

```python
ints, bins, patches = plt.hist(priceperlb, normed=True)
mids = [(bins[i] + bins[i + 1]) / 2 for i in range(k)]
rel_frequencies_group = np.divide(ints, n)

norm_rv = sts.norm(np.mean(priceperlb), np.std(priceperlb))
x_values = np.linspace(0,max_value,100)
norm_pdf = norm_rv.pdf(x_values)
plt.plot(x_values, norm_pdf, color="red")

plt.show()
```

```python
pd.DataFrame({"n":ints, "$x_a, x_b$":["[ {} ; {} ]"
                                   .format(bins[i], bins[i + 1]) for i in range(k)],
              "$w_i$":rel_frequencies_group, "$x_i$":mids})
```

Вычислим основные характеристики:

```python
expectancy = np.mean(priceperlb)
variance = np.var(priceperlb)
standard_deviation = np.std(priceperlb)
median = np.median(priceperlb)
print("Мат.ожидание: {}\nДисперсия: {}\nСтанд.отклонение: {}\nМедиана: {}\n"
      .format(expectancy, variance, standard_deviation, median))
```

Проведём анализ разбиения на группы, для чего вычислим:
    
Межгрупповую дисперсию:

$$\sigma_{мг}^{2} = \frac{\sum_{i=1}^{k} (\tilde{x_i} - \tilde{x_{общ}})^{2} * f_i}{\sum_{i=1}^{k}f_i}$$

```python
mean = np.mean(priceperlb)
#межгрупповая дисперсия
intergroup_var = sum([((np.mean(priceperlb[np.where((priceperlb > bins[i]) & (priceperlb <= bins[i + 1]))]) 
                    - mean) ** 2) * rel_frequencies_group[i] for i in range(k)]) / sum(rel_frequencies_group)
print("Межгрупповая дисперсия: {}".format(intergroup_var))
```

Внутригрупповые дисперсии:

$$\sigma_{j}^{2} = \frac{\sum_{i=1}^{n}(x_i - \tilde{x_j})^{2}}{n}$$

```python
#средняя из внутригрупповых дисперсий
ingroup_var = np.mean([np.var(priceperlb[np.where((priceperlb > bins[i]) & (priceperlb <= bins[i + 1]))])
                       for i in range(k)])
print("Средняя из внутригрупповых дисперсий: {}".format(ingroup_var))
```

По правилу сложения дисперсий, дисперсия равна сумме средней из внутригрупповых дисперсий и межгрупповой дисперсии:
$$\sigma^{2} = \sigma_{мг}^{2} + \tilde{\sigma^{2}}$$

```python
print("Дисперсия: {}\nСумма: {}".format(variance, ingroup_var + intergroup_var))
print("Различие: {} %".format(np.abs(variance - (ingroup_var + intergroup_var)) / variance * 100))
```

А также функцию распределения для $x_i$:

```python
cum_frequencies = [sum(rel_frequencies_group[0:index]) for index, value in enumerate(rel_frequencies_group)]
plt.plot(mids, cum_frequencies, color='red')
plt.title("График функции распределения:")
plt.xlabel("x(i)")
plt.ylabel("F(x)")
plt.show()
```

Исходя из визуального представления функций распределения и частот выдвинем гипотезу, что данное распределение является нормальным.

Теоретические частоты:


$$m' = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x - \mu)^{2}}{2\sigma^{2}}}$$

```python
theor_frequencies = [(1.0 / (standard_deviation * np.sqrt(2*np.pi)) *
                     np.exp(-1 * ((x - mean) ** 2) / (2 * variance))) for x in mids]
```

Для проверки вычислим критерий Пирсона:

$$\chi^{2} = \sum\frac{(m_i - m_i^{'})^{2}}{m_i^{'}}$$

```python
xi_squared = sum([((rel_frequencies_group[i] - theor_frequencies[i]) ** 2) 
                  / theor_frequencies[i] for i in range(len(rel_frequencies_group)) ])
print("Критерий Пирсона: {}".format(xi_squared))
```

Полученное значение меньше табличного, значит, гипотеза не опровергается.

Критерий Романовского находим по формуле:

$$R = \frac{|\chi^{2} - k|}{\sqrt{2k}}$$

```python
R = np.abs(xi_squared - k) / np.sqrt(2 * k)
print("Критерий Романовского: {}".format(R))
```

Полученное значение меньше 3, значит, гипотеза не опровергается.

Вычислим значение критерия Ястремского:

$$J = \frac{|\chi^{2} - k|}{\sqrt{2k + 2,4}}$$

```python
J = np.abs(xi_squared - k) / np.sqrt(2 * k + 2.4)
print("Критерий Ястремского: {}".format(J))
```

Полученный результат меньше 3, значит, гипотеза не опровергается.

Таким образом, на основании критериев Пирсона, Романовского и Ястремского можно сделать вывод, что гипотеза подтверждается и данное распределение является нормальным.
